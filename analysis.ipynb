```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from datetime import datetime 
```

```python
#loading dataset 
df = pd.read_csv('healthcare_dataset.csv')
print("Columns in the dataset:", df.columns.tolist())
df.head(5)
```

```python
print("Dataset Information:")
print(df.info())

#checks for any missing or null values
print("Missing values:")
print(df.isnull().sum())

#summary statistics for numeric columns 
print("Basic statistics:")
df.describe()
```

```python
#Fix date columns 

df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])

df["Length of stay"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days

print("New columns added after fixing date columns")
print(df[["Date of Admission", "Discharge Date", "Length of stay"]].head())
```

```python
plt.figure(figsize=(10,5))

condition_count = df["Medical Condition"].value_counts()

sns.barplot(x=condition_count.values, y=condition_count.index, palette="flare")
plt.title("Common medical conditions")
plt.xlabel("Number of patients")
plt.ylabel("Medical condition")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.countplot(x="Admission Type", data= df, palette="flare")
plt.title("Admission type distribution")
plt.xlabel("Admission type")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Medical Condition", y="Age", data=df, palette="flare")

plt.title("Age distribution by medical condition")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(df["Billing Amount"], bins=30, color="skyblue", kde=True)

plt.title("Billing amount distribution")
plt.xlabel("Billing Amount")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x='Admission Type', y='Length of stay', data=df, palette='flare')
plt.title('Length of stay by admission type')
plt.xlabel("Admission type")
plt.ylabel("days")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

blood_count = df["Blood Type"].value_counts()
sns.barplot(x=blood_count.index, y=blood_count.values, palette='flare')
plt.title("Patient blood type distribution")
plt.xlabel("Blood type")
plt.ylabel("Count")
plt.show()
```

```python
print("KEY INSIGHTS:")
print("Total patients:", len(df))
print("Average age of patients:", df["Age"].mean(), "years")
print("Average length of stay:", df["Length of stay"].mean(), "days")
print("Average billing amount:", df["Billing Amount"].mean(), "USD" )
print("Most common condition:", df["Medical Condition"].mode()[0])
print("Most common admission type:", df["Admission Type"].mode()[0])
print("Most common blood type:", df["Blood Type"].mode()[0])
```
